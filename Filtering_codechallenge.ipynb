```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
import scipy
import scipy.io as sio
import scipy.io.wavfile
import copy
```

```python
# load data

rawdata = sio.loadmat('filtering_codeChallenge.mat')

data  = np.squeeze(rawdata['x'])
fs = rawdata['fs'][0]
filtered_data = np.squeeze(rawdata['y'])

# time vector
pnts = len(data)
time = np.arange(0,pnts)/fs
time = time.T

# compute power spectrum and frequencies vector
pwr = np.abs(scipy.fft.fft(data)/pnts)**2
ylimit = max(pwr)
filtered_pwr = np.abs(scipy.fft.fft(filtered_data)/pnts)**2

hz  = np.linspace(0,fs,pnts)

### plotting
# time-domain signal
plt.subplot(121)
plt.figure(figsize=(20,5))
plt.plot(time[0:-1:],data[0:-1:],'k',label='Original')
plt.plot(time[0:-1:],filtered_data[0:-1:],'r',label='filtered')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Time domain')
plt.legend()
plt.show()
# plot power spectrum
plt.subplot(122)
plt.figure(figsize=(20,5))
plt.plot(hz,pwr,'k',label='Original')
plt.plot(hz,filtered_pwr,'r',label='Filtered')

plt.xlim([0,fs/2])
plt.ylim([0,ylimit*1.2])
plt.xlabel('Frequency (Hz)')
plt.ylabel('Power')
plt.title('Frequency domain')

```
